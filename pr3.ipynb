Theory :


1. What is data wrangling? Difference between data wrangling and data preprocessing ?

ANS :
Data Wrangling: The process of cleaning, transforming, and organizing raw data into a usable format for analysis.
Difference:
Data Wrangling → Broader process (cleaning + transforming + structuring data).
Data Preprocessing → A part of wrangling, mainly focused on cleaning (handling missing values, normalization, encoding).


2. Importance of Exploratory Data Analysis (EDA) in Pandas ?

ANS :
EDA helps to:
Understand data structure and patterns
Detect missing values and outliers
Identify relationships between variables
Make better decisions before applying models.


3. Differences between Series, DataFrame, and Panel ?
ANS :
Series: 1D labeled array (like a single column)
DataFrame: 2D labeled data (rows and columns like a table)
Panel: 3D data structure (collection of DataFrames) (Note: Panel is deprecated in modern pandas).


4. Explain the use of:
a. isnull()
b. notnull()
c. dropna()
d. fillna() .
ANS :
a. isnull() → Checks for missing values (returns True if null)
b. notnull() → Opposite of isnull (True if value exists)
c. dropna() → Removes rows/columns with missing values
d. fillna() → Replaces missing values with a specified value


5. What is column transformation and why is it important in data analysis ?
ANS:
Column Transformation: Modifying column data (e.g., scaling, encoding, formatting).
Importance:
Improves data quality
Makes data suitable for analysis or machine learning
Ensures consistency and accuracy .



```python
import pandas as pd

# Step 1: Create dataset (no need to upload file)
data = {
    'Name': ['Amit','Neha','Ravi','Priya','Amit','Neha','Ravi','Priya'],
    'Subject': ['Math','Math','Science','Science','Math','Math','Science','Science'],
    'Semester': [1,1,1,1,2,2,2,2],
    'Marks': [78,85,82,88,81,79,85,90]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# ---------------- BASIC OPERATIONS ----------------

# 1. Select a single column
print("\n1. Selecting 'Marks' column:")
print(df['Marks'])

# 2. Add a new column with constant values
df['College'] = "ABC University"
print("\n2. After adding new column 'College':")
print(df)

# 3. Sort DataFrame by one column
sorted_df = df.sort_values(by='Marks', ascending=False)
print("\n3. Sorted DataFrame by Marks:")
print(sorted_df)

# 4. Rename DataFrame columns
df = df.rename(columns={'Marks': 'Score'})
print("\n4. After renaming 'Marks' to 'Score':")
print(df)

# 5. Display summary statistics
print("\n5. Summary Statistics:")
print(df.describe())

# ---------------- DATA WRANGLING (MAIN TASK) ----------------

# Group by Subject and Semester and calculate average
grouped = df.groupby(['Subject','Semester'])['Score'].mean()

print("\nGrouped Average (Subject & Semester):")
print(grouped)

# Convert to DataFrame
summary = grouped.reset_index()

print("\nFinal Summary DataFrame:")
print(summary)

# Pivot Table
pivot = df.pivot_table(values='Score', index='Subject', columns='Semester', aggfunc='mean')

print("\nPivot Table:")
print(pivot)

# Save result
summary.to_csv('final_summary.csv', index=False)
```

Conclusion :


The experiment successfully demonstrated basic data manipulation and data wrangling using pandas. Operations such as selecting columns, adding new columns, sorting data, renaming columns, and generating summary statistics were performed effectively. Furthermore, the dataset was grouped by subject and semester to calculate average scores, providing meaningful insights into student performance. The use of pivot tables helped in presenting the data in a structured and easy-to-understand format. Overall, pandas proved to be a powerful tool for organizing, analyzing, and summarizing data efficiently.
