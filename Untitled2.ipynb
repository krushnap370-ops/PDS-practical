```python
MAX = 1000
MIN = -1000

def minimax(depth, index, isMax, values, alpha, beta):
    if depth == 3:      # leaf
        return values[index]

    if isMax:
        best = MIN

        best = max(best, minimax(depth+1, index*2, False, values, alpha, beta))
        alpha = max(alpha, best)

        if beta <= alpha:   # prune
            return best

        best = max(best, minimax(depth+1, index*2+1, False, values, alpha, beta))
        alpha = max(alpha, best)

        return best

    else:
        best = MAX

        best = min(best, minimax(depth+1, index*2, True, values, alpha, beta))
        beta = min(beta, best)

        if beta <= alpha:   # prune
            return best

        best = min(best, minimax(depth+1, index*2+1, True, values, alpha, beta))
        beta = min(beta, best)

        return best


values = [3, 5, 6, 9, 1, 2, 0, -1]
print("Optimal value:", minimax(0, 0, True, values, MIN, MAX))
```
