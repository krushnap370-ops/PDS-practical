Q1. what is data ingestion and why is it important in data science ? Which python
library is commonly used for data ingestion and cleaning ?
Ans: Data ingestion is the process of collecting, importing and transferring data
from various source into a system where it can be a stored , processed and
analyzed. These source can include :
1. Databases
2. CSV/Excel files
3. APIs
4. Web scraping
5. Streaming data
Data ingestion can be :
1. Batch ingestion - Data is collected and processed in chunks at
scheduled intervals .
2. Real-time ingestion -Data is continuously imported as it is generated .
Importance in data science : Data ingestion is critical because :
1. Foundation of analysis -data science begins with data .Without properly
ingested data analysis cannot happen .
2. ensure data availability - It makes raw data accessible.

Q4. What is the difference between head() and tail() function in pandas ?

Ans : head() :
1. return the first n row of the dataset.
2. Use this to check column headers , data types and ensures the first few
rows align with your expectation.
tail():
1. Returns the first n rows of the data set
2. Use this to check the total height of your dataset or to see if there are
empty rows at the very end of a file .


Q5. How can you identify missing values in a dataset?

Ans: Missing values can be identified by checking for null , Nan , blank cells or
special symbils in the data set. Tis can be done using function such as isnull() in
python , ISBLANK() or COUNTBLANK() in Excel, or is NULL in SQL to detect and
count missing data .

```python
import pandas as pd

# 1. Load Excel file
df = pd.read_excel("data.csv.xlsx")

# Display first five rows
print("First 5 rows:")
print(df.head())

# 2. Identify missing values
print("\nMissing values:")
print(df.isnull())

# Count missing values
print("\nMissing values count:")
print(df.isnull().sum())

# 3. Replace missing values with 0
df_filled = df.fillna(0)
print("\nAfter replacing missing values with 0:")
print(df_filled)

# 4. Rename column names to lowercase
df_filled.columns = df_filled.columns.str.lower()
print("\nColumn names in lowercase:")
print(df_filled.columns)

# 5. Display dataset shape
print("\nDataset shape (rows, columns):")
print(df_filled.shape)
```
